```python
import gym
class Agent:
    def __init__(self,env):
        self.actionNumber=env.action_space.n
        
    def getAction(self, state):
        angle=state[2]
        if (angle<0):
            return 0
        elif (angle>0):
            return 1
```

```python
env = gym.make('CartPole-v0')
a=Agent(env)
for j in range(10):
    state=env.reset()
    for i in range(10000):
        state, reward, done, info=env.step(a.getAction(state)) # take a random action
        env.render()
        print(state)
        if (done):
            print("No .of steps "+str(i))
            break
            
env.close()
```
